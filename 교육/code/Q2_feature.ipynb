# 로그 데이터 Enter/Quit 고아(?) 찾기

```python
import pandas as pd
import zipfile
import os

# [설정] ZIP 파일 경로
ZIP_PATH = '/content/drive/MyDrive/DSL_EDA_EDU/KT4.zip'

def find_orphan_logs(df):
    """
    유저 데이터프레임에서 Enter/Quit 짝이 안 맞는 '고아'가 있는지 검사
    반환값: 고아 발견 여부 (True/False), 상세 내용 (String)
    """
    # 1. 해설(Explanation) 아이템만 필터링
    expl_df = df[df['item_id'].str.startswith('e', na=False)].copy()

    # 2. 시간순 정렬
    expl_df = expl_df.sort_values('timestamp')

    # 3. 아이템별로 그룹화하여 검사
    orphan_details = []
    has_orphan = False

    # 각 아이템(예: e1234)별로 enter/quit 흐름 확인
    for item_id, group in expl_df.groupby('item_id'):
        actions = group['action_type'].tolist()

        # (1) 단순 개수 비교
        n_enter = actions.count('enter')
        n_quit = actions.count('quit')

        if n_enter != n_quit:
            has_orphan = True
            orphan_details.append(f"[{item_id}] 개수 불일치 (Enter: {n_enter}, Quit: {n_quit})")
            continue # 개수부터 다르면 더 볼 것도 없음

        # (2) 순서 검증 (Stack 활용)
        # 올바른 순서: Enter가 나오면 스택에 넣고, Quit이 나오면 뺀다.
        # Quit이 나왔는데 스택이 비어있으면 -> Quit이 고아
        # 끝났는데 스택이 남아있으면 -> Enter가 고아
        stack = 0
        for action in actions:
            if action == 'enter':
                stack += 1
            elif action == 'quit':
                stack -= 1

            if stack < 0: # Quit이 먼저 나옴
                has_orphan = True
                orphan_details.append(f"[{item_id}] Quit이 먼저 등장함 (순서 오류)")
                break

        if stack > 0: # Enter가 닫히지 않음
            has_orphan = True
            orphan_details.append(f"[{item_id}] 닫히지 않은 Enter 존재 (순서 오류)")

    return has_orphan, orphan_details

def scan_zip_for_orphans(zip_path, limit=None):
    """
    ZIP 파일을 돌며 고아가 있는 유저를 찾아냄
    limit: 테스트용으로 몇 명만 검사할지 (None이면 전체)
    """
    orphan_users = []

    with zipfile.ZipFile(zip_path, 'r') as z:
        csv_files = [f for f in z.namelist() if f.endswith('.csv')]
        print(f"총 {len(csv_files)}명의 유저에 대해 무결성 검사를 시작합니다.")

        for idx, filename in enumerate(csv_files):
            if limit and idx >= limit: break

            try:
                with z.open(filename) as f:
                    df = pd.read_csv(f)
                    if df.empty: continue

                    user_id = os.path.splitext(os.path.basename(filename))[0]

                    # 고아 찾기 함수 실행
                    is_bad, details = find_orphan_logs(df)

                    if is_bad:
                        orphan_users.append({
                            'user_id': user_id,
                            'issues': str(details[:3]) + ('...' if len(details)>3 else '') # 너무 길면 자름
                        })

            except Exception as e:
                print(f"Error {filename}: {e}")

            if (idx + 1) % 5000 == 0:
                print(f"{idx+1}명 검사 완료...")

    return pd.DataFrame(orphan_users)

# --- 실행 ---
orphan_report = scan_zip_for_orphans(ZIP_PATH) # 전체 검사 시 시간 소요 주의

if not orphan_report.empty:
     print(f"⚠️ 총 {len(orphan_report)}명의 유저에게서 고아 로그가 발견되었습니다!")
     print(orphan_report.head())
else:
     print("✅ 고아 로그가 하나도 없습니다! 데이터가 완벽합니다.")
```

# **Feature 함수**

```python
# feature 함수
def calculate_average_KT4_features(df, threshold=1800):
  """
  """
  ## 하루 평균 학습시간 part

  # 전처리: timestamp -> datetime  date
  df['datetime'] = pd.to_datetime(df['timestamp'], unit='ms')
  df['date'] = df['datetime'].dt.date

  # 시간순 정렬(혹시나)
  df = df.sort_values(by='datetime')

  # (1) 행동간 시차 계산 (단위: 초)
  df['duration'] = (df['datetime']-df['datetime'].shift(1)).dt.total_seconds()

  # (2) 유효 학습 시간 필터링
  # duration이 Nan이 아니고
  # 0보다 크며
  # threshold 이하인 경우 제외
  valid_duration = df[
      (df['duration'].notna()) &
      (df['duration'] > 0) &
      (df['duration'] <= threshold)
  ]

  # (3) 총 학습시간 (초)
  total_study_time = valid_duration['duration'].sum()

  # (4) 총 학습일수
  total_study_days = df['date'].nunique()

  # (5) 하루 평균 학습시간 계산
  daily_avg_study_time = total_study_time / total_study_days if total_study_days > 0 else 0

  ## 최대 연속학습일수 part

  # (1). 날짜 추출 및 내림차순 정렬
  unique_dates = sorted(df['date'].unique(), reverse=True) # df.unique() -> df['date'].unique() 수정
  if not unique_dates:
    return {
      'total_logs' : len(df),
      'total_study_time' : 0,
      'total_study_days' : 0,
      'daily_avg_study_time' : 0,
      'maxhot' : 0,
      'erase_choice_count' : 0,
      'explanation_choice_count' : 0,
      'total_explanation_duration_sec' : 0,
      'adaptive_offer_count' : 0,
      'total_adaptive_duration_sec' : 0,
  } # unique_dates가 비어있을 경우 딕셔너리 반환하도록 수정

  hot = 1 ; maxhot = 1
  # (2). 최신 날짜부터 과거로 가며 차이가
  for i in range(len(unique_dates)-1):
    # (오늘-어제) 의 차이가 1일이면 연속
    if (unique_dates[i] - unique_dates[i+1]).days == 1:
      hot += 1
    else:
      hot = 1
    if hot > maxhot:
      maxhot = hot

  ## 선택지 제거활용 part
  # df의 action_type에서 erase_choic의 횟수를 세기
  erase_choice_count = df[df['action_type'] == 'erase_choice'].shape[0]

  ## 해설 채택수 part
  # df의 item_id가 e로 시작하는 것 세기
  explanation_choice_count = df[df['item_id'].str.startswith('e')].shape[0]

  ## 해설 체류시간
  # df의 item_id에서 e로 시작하는 것의 enter와 quit의 사이 시간을 모두 합하기
  """
  한 유저의 DataFrame을 받아 해설(e로 시작하는 item)에 머무른 총 시간을 계산 (단위: 초)
  """
  # (1). 'e'로 시작하는 item_id만 필터링 (Explanation)
  # na=False는 NaN 값이 있을 경우 오류 방지
  expl_df = df[df['item_id'].str.startswith('e', na=False)].copy()

  # (2). 'enter'와 'quit' 행동만 남기기
  # play_audio, pause_audio 등 중간 행동은 체류 시간 계산(시작~끝)에는 필요 없음
  expl_df = expl_df[expl_df['action_type'].isin(['enter', 'quit'])]

  # (3). 시간순 정렬 (필수: 꼬여있으면 로직이 깨짐)
  expl_df = expl_df.sort_values('timestamp')

  # (4). Shift를 이용해 '이전 행'의 정보 가져오기 (Vectorized Operation)
  # 한 줄씩 for문 돌리는 것보다 훨씬 빠름
  expl_df['prev_action'] = expl_df['action_type'].shift(1)
  expl_df['prev_item'] = expl_df['item_id'].shift(1)
  expl_df['prev_timestamp'] = expl_df['timestamp'].shift(1)

  # (5). 유효한 '체류 구간' 찾기 (조건 3가지 만족 시)
  # 1) 현재 행이 'quit'이고
  # 2) 바로 이전 행이 'enter'이며
  # 3) 아이템 ID가 서로 같아야 함 (e1 enter -> e1 quit)
  valid_mask = (
        (expl_df['action_type'] == 'quit') &
        (expl_df['prev_action'] == 'enter') &
        (expl_df['item_id'] == expl_df['prev_item'])
    )

  # (6). 조건에 맞는 행들만 뽑아서 시간 차이 계산
  valid_logs = expl_df[valid_mask]

  # EdNet의 timestamp는 밀리초(ms) 단위이므로 1000으로 나눠서 초(sec)로 변환
  # (datetime 객체끼리 빼는 것보다 원본 timestamp(int)끼리 빼는 게 더 빠르고 간편합니다)
  if valid_logs.empty:
        total_explanation_duration_sec = 0
  else:
        duration_ms = (valid_logs['timestamp'] - valid_logs['prev_timestamp']).sum()
        total_explanation_duration_sec = duration_ms / 1000

  ## 추천 채택수
  # df의 source가 adaptive_offer인 것 세기
  adaptive_offer_count = df[df['source'] == 'adaptive_offer'].shape[0]

  ## 추천 체류시간
  # df의 source가 adaptive_offer인 것의 enter와 quit의 사이 시간을 모두 합하기
  """
  한 유저의 DataFrame을 받아 추천 콘텐츠(adaptive_offer)에 머무른 총 시간을 계산 (단위: 초)
  """
  # 1. source가 'adaptive_offer'인 것만 필터링
  # na=False는 NaN 값이 있을 경우 오류 방지
  adaptive_df = df[df['source'] == 'adaptive_offer'].copy()

  # 2. 'enter'와 'quit' 행동만 남기기
  # 중간 행동(play, pause 등)은 체류 시간 계산(시작~끝)에는 필요 없음
  adaptive_df = adaptive_df[adaptive_df['action_type'].isin(['enter', 'quit'])]

  # 3. 시간순 정렬 (필수: 꼬여있으면 로직이 깨짐)
  adaptive_df = adaptive_df.sort_values('timestamp')

  # 4. Shift를 이용해 '이전 행'의 정보 가져오기 (Vectorized Operation)
  adaptive_df['prev_action'] = adaptive_df['action_type'].shift(1)
  adaptive_df['prev_item'] = adaptive_df['item_id'].shift(1)
  adaptive_df['prev_timestamp'] = adaptive_df['timestamp'].shift(1)

  # 5. 유효한 '체류 구간' 찾기 (엄격한 짝 맞추기)
  # (1) 현재 행이 'quit'이고
  # (2) 바로 이전 행이 'enter'이며
  # (3) 아이템 ID가 서로 같아야 함 (같은 추천 콘텐츠 내에서의 활동)
  valid_mask = (
        (adaptive_df['action_type'] == 'quit') &
        (adaptive_df['prev_action'] == 'enter') &
        (adaptive_df['item_id'] == adaptive_df['prev_item'])
    )

  # 6. 조건에 맞는 행들만 뽑아서 시간 차이 계산
  valid_logs = adaptive_df[valid_mask]

  # EdNet의 timestamp는 밀리초(ms) 단위이므로 1000으로 나눠서 초(sec)로 변환
  if valid_logs.empty:
        total_adaptive_duration_sec = 0
  else:
        duration_ms = (valid_logs['timestamp'] - valid_logs['prev_timestamp']).sum()
        total_adaptive_duration_sec = duration_ms / 1000

  return {
      'total_logs' : len(df),
      'total_study_time' : total_study_time,
      'total_study_days' : total_study_days,
      'daily_avg_study_time' : daily_avg_study_time,
      'maxhot' : maxhot,
      'erase_choice_count' :erase_choice_count,
      'explanation_choice_count' : explanation_choice_count,
      'total_explanation_duration_sec' : total_explanation_duration_sec,
      'adaptive_offer_count' : adaptive_offer_count,
      'total_adaptive_duration_sec' : total_adaptive_duration_sec,

  }

def process_zip_for_KT4(zip_path):
    results = []

    with zipfile.ZipFile(zip_path, 'r') as z:
        csv_files = [f for f in z.namelist() if f.endswith('.csv')]
        total_files = len(csv_files)

        print(f"총 {total_files}개의 유저 데이터를 처리합니다.")

        for idx, filename in enumerate(csv_files):
            try:
                with z.open(filename) as f:
                    # (1) 데이터 로드
                    df = pd.read_csv(f)

                    if df.empty:
                        continue

                    # (2) 유저 ID 추출 (파일명 u123.csv -> u123)
                    user_id = os.path.splitext(os.path.basename(filename))[0]

                    # (3) 함수 적용 (전처리 및 Feature 추출)
                    feats = calculate_average_KT4_features(df)

                    # (4) 유저 ID 추가 후 결과 리스트에 저장
                    feats['user_id'] = user_id
                    results.append(feats)

            except Exception as e:
                print(f"Error in {filename}: {e}")

            # 진행 상황 모니터링 (1000개마다 출력)
            if (idx + 1) % 1000 == 0:
                print(f"[{idx+1}/{total_files}] 처리 완료...")

    # ---------------------------------------------------------
    # [3] 최종 결과 DataFrame 생성
    # ---------------------------------------------------------
    final_df = pd.DataFrame(results)
    return final_df
```

```python
## 학습비중 추출 함수
def extract_source_features(df):
    """
    유저 DataFrame에서 Source별 비중(Ratio)을 추출
    """
    # 1. 전체 로그 수
    total_logs = len(df)

    # 2. Source별 비율 계산 (normalize=True로 하면 0~1 사이 비율이 나옴)
    # dropna=False를 해야 NaN(결측치)도 비율에 포함됨 (필요시)
    source_ratios = df['source'].value_counts(normalize=True)

    # 결과 딕셔너리 생성
    result = {'user_id': None, 'total_logs': total_logs}

    # 3. 주요 Source들에 대해 비율 저장 (없으면 0.0)
    for src in TARGET_SOURCES:
        # 컬럼명 예: ratio_sprint, ratio_adaptive_offer
        result[f'ratio_{src}'] = source_ratios.get(src, 0.0)

    return result

def process_zip_for_sources(zip_path):
    results = []
    with zipfile.ZipFile(zip_path, 'r') as z:
        csv_files = [f for f in z.namelist() if f.endswith('.csv')]

        print(f"총 {len(csv_files)}명의 유저 데이터를 분석합니다.")

        for idx, filename in enumerate(csv_files):
            try:
                with z.open(filename) as f:
                    df = pd.read_csv(f)
                    if df.empty: continue

                    # (1) 소스 컬럼이 있는지 확인
                    if 'source' not in df.columns:
                        continue

                    user_id = os.path.splitext(os.path.basename(filename))[0]

                    # (2) Feature 추출 함수 실행
                    feats = extract_source_features(df)
                    feats['user_id'] = user_id

                    results.append(feats)

            except Exception as e:
                print(f"Error {filename}: {e}")

            if (idx + 1) % 5000 == 0:
                print(f"{idx+1}명 처리 완료...")

    return pd.DataFrame(results)
```

```python
def calculate_average_KT1_features(df):
  """
  KT1의 평균 정답률과 평균 응답시간을 계산
  """
  # 평균 정답률 구하기
  average_accuracy = df['correct'].mean()

  # 평균 응답시간 구하기
  average_response_time_ms = df['elapsed_time'].mean()

  return {
      'average_accuracy': average_accuracy,
      'average_response_time_ms': average_response_time_ms}

def process_zip_for_KT1(zip_path):
    results = []

    with zipfile.ZipFile(zip_path, 'r') as z:
        csv_files = [f for f in z.namelist() if f.endswith('.csv')]
        total_files = len(csv_files)

        print(f"총 {total_files}개의 유저 데이터를 처리합니다.")

        for idx, filename in enumerate(csv_files):
            try:
                with z.open(filename) as f:
                    # (1) 데이터 로드
                    df = pd.read_csv(f)

                    if df.empty:
                        continue

                    # (2) 유저 ID 추출 (파일명 u123.csv -> u123)
                    user_id = os.path.splitext(os.path.basename(filename))[0]

                    # (3) 함수 적용 (전처리 및 Feature 추출)
                    feats = calculate_average_KT1_features(df)

                    # (4) 유저 ID 추가 후 결과 리스트에 저장
                    feats['user_id'] = user_id
                    results.append(feats)

            except Exception as e:
                print(f"Error in {filename}: {e}")

            # 진행 상황 모니터링 (1000개마다 출력)
            if (idx + 1) % 1000 == 0:
                print(f"[{idx+1}/{total_files}] 처리 완료...")

    # ---------------------------------------------------------
    # [3] 최종 결과 DataFrame 생성
    # ---------------------------------------------------------
    final_df = pd.DataFrame(results)
    return final_df
```
